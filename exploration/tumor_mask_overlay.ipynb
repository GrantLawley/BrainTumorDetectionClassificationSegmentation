```python
import numpy as np
import matplotlib.pyplot as plt
import torchvision.transforms.functional as TF
import torchvision.io as io

# --- Paths ---
image_path = r"..\brisc2025\segmentation_task\train\images\brisc2025_train_00007_gl_ax_t1.jpg"
mask_path  = r"..\brisc2025\segmentation_task\train\masks\brisc2025_train_00007_gl_ax_t1.png"
new_size   = [224, 224]
alpha      = 0.45

image = io.decode_image(image_path)[0].unsqueeze(0).float() / 255.0
mask  = io.decode_image(mask_path)[0].unsqueeze(0).int()
mask = (mask / 255.0).round()

image_resized = TF.resize(image, new_size)
mask_resized  = TF.resize(mask, new_size, interpolation=TF.InterpolationMode.NEAREST_EXACT)

def overlay(img_np, mask_np, alpha=0.45, color=(1, 0, 0)):
    rgb = np.stack([img_np] * 3, axis=-1).clip(0, 1)
    tumour = mask_np > 0
    for i, c in enumerate(color):
        rgb[tumour, i] = (1 - alpha) * rgb[tumour, i] + alpha * c
    return rgb

img_np          = image.squeeze().numpy()
mask_np         = mask.squeeze().numpy()
img_resized_np  = image_resized.squeeze().numpy()
mask_resized_np = mask_resized.squeeze().numpy()

# --- Plot separately ---
dpi = 100

# Original
h_orig, w_orig = img_np.shape
fig1, ax1 = plt.subplots(figsize=(w_orig / dpi, h_orig / dpi), dpi=dpi)
ax1.imshow(overlay(img_np, mask_np, alpha))
ax1.set_title(f"Original ({w_orig}x{h_orig})")
ax1.axis("off")
plt.tight_layout()
plt.show()

# Resized
h_new, w_new = img_resized_np.shape
fig2, ax2 = plt.subplots(figsize=(w_new / dpi, h_new / dpi), dpi=dpi)
ax2.imshow(overlay(img_resized_np, mask_resized_np, alpha))
ax2.set_title(f"Resized ({w_new}x{h_new})")
ax2.axis("off")
plt.tight_layout()
plt.show()
```
